Descargamos un video pequeño:

```python
import pytube as pt


# IlloJuan y Masi responden la pregunta más importante de su vida: ¿Sois más de perros o gatos?
# 10 min
yt = pt.YouTube("https://www.youtube.com/watch?v=_3tb3ZnWffM") 
# yt = pt.YouTube("https://www.youtube.com/watch?v=LAa4e9H1Nlo") 
video_title = yt.title
stream = yt.streams.filter(only_audio=True).first()
stream.download(filename=f"{video_title}.mp3")
audio_path = f"/kaggle/working/{video_title}.mp3"
```

Transcribimos el audio con whisper: 
- En este caso se usa el modelo `small`. Podemos considerar usar modelos mas pequeños para optimizar la velocidad de inferencia.
- La transcrición es buena pero tiene algunos errores.
- Tardo 2 min en transcribir un video de 10 min.
- Guardo la transcripcion en un archivo `.txt`

```python
%%time 
!python -m whisper "$audio_path" --verbose True --output_format txt --output_dir . --language Spanish
```
